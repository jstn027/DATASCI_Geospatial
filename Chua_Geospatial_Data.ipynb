```python
fig = plt.figure(figsize=(30, 20))
ax = fig.add_subplot(111)
ph.plot(ax=ax, color='white', edgecolor='dimgray')
healthsites.plot(ax=ax, column="amenity", alpha=0.35, legend=True)
ax.set_title('Health-related Amenities in the Philippines', fontsize=15)
ax.axis('off');
```

```python
healthsites['amenity'].value_counts()
```

```python
data_pharmacy= pd.DataFrame()
data_pharmacy= data_pharmacy.append(healthsites.loc[healthsites['amenity']=='pharmacy'])
data_pharmacy= data_pharmacy.reset_index(drop=True)

data_clinic= pd.DataFrame()
data_clinic= data_clinic.append(healthsites.loc[healthsites['amenity']=='clinic'])
data_clinic= data_clinic.reset_index(drop=True)

data_hospital= pd.DataFrame()
data_hospital= data_hospital.append(healthsites.loc[healthsites['amenity']=='hospital'])
data_hospital= data_hospital.reset_index(drop=True)

data_dentist= pd.DataFrame()
data_dentist= data_dentist.append(healthsites.loc[healthsites['amenity']=='dentist'])
data_dentist= data_dentist.reset_index(drop=True)

data_doctor= pd.DataFrame()
data_doctor= data_doctor.append(healthsites.loc[healthsites['amenity']=='doctors'])
data_doctor= data_doctor.reset_index(drop=True)

data_unkown= pd.DataFrame()
data_unkown= data_unkown.append(healthsites.loc[healthsites['amenity']==''])
data_unkown= data_unkown.reset_index(drop=True)

data_laboratory= pd.DataFrame()
data_laboratory= data_laboratory.append(healthsites.loc[healthsites['amenity']=='laboratory'])
data_laboratory= data_laboratory.reset_index(drop=True)

data_socialfacility= pd.DataFrame()
data_socialfacility= data_socialfacility.append(healthsites.loc[healthsites['amenity']=='social_facility'])
data_socialfacility= data_socialfacility.reset_index(drop=True)

data_healthcare= pd.DataFrame()
data_healthcare= data_healthcare.append(healthsites.loc[healthsites['amenity']=='healthcare'])
data_healthcare= data_healthcare.reset_index(drop=True)
```

```python
mm = ph[ph.NAME_1=="Metropolitan Manila"]

mm_within = gpd.sjoin(data_pharmacy, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="viridis", alpha=0.35, legend=True)
ax.set_title('Pharmacies in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_clinic, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="spring", alpha=0.35, legend=True)
ax.set_title('Clinics in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_hospital, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="summer", alpha=0.35, legend=True)
ax.set_title('Hospitals in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_dentist, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="autumn", alpha=0.35, legend=True)
ax.set_title('Dentists in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_doctor, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="Wistia", alpha=0.35, legend=True)
ax.set_title('Doctors in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_unkown, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="cool", alpha=0.35, legend=True)
ax.set_title('Unlabeled in Metro Manila', fontsize=15)
ax.axis('off');

mm_within = gpd.sjoin(data_healthcare, mm, op='within')
mm_within= mm_within.reset_index(drop=True)
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)
mm.plot(ax=ax, color='white', edgecolor='dimgray')
mm_within.plot(ax=ax, column="amenity", cmap="gist_heat", alpha=0.35, legend=True)
ax.set_title('Healthcare in Metro Manila', fontsize=15)
ax.axis('off');
```
